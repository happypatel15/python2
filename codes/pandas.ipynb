<h1> Pandas Series </h1>
A Pandas Series is like a column in a table.<br>
It is a one-dimensional array holding data of any type.

```python
import pandas as pd

a = [1,2,3]

print(pd.Series(a))
```

```python
print(pd.Series(a,index=["x","y","z"]))
```

```python
a = (1,2,3)
print(pd.Series(a,index=["x","y","z"]))
```

```python
a = {"a":1,"b":2}
print(pd.Series(a,index=["1","2"]))
print(pd.Series(a,index=["a","b"]))
```

<h3> Data Frames </h3>

```python
data = {
    "calories" : [420,380,390],
    "dration" : [50,40,45]
}

print(pd.DataFrame(data))
print()
print(pd.DataFrame(data,index=["x","z","z"]))
print()
print(pd.DataFrame(data)["calories"])
print()
print(pd.DataFrame(data)["calories"][0])
print()
print(pd.DataFrame(data).loc[2])
print()
print(pd.DataFrame(data).loc[[0,2]])
```
