<h1> Pandas Series </h1>
A Pandas Series is like a column in a table.<br>
It is a one-dimensional array holding data of any type.

```python
import pandas as pd

a = [1,2,3]

print(pd.Series(a))
```

```python
print(pd.Series(a,index=["x","y","z"]))
```

```python
a = (1,2,3)
print(pd.Series(a,index=["x","y","z"]))
```

```python
a = {"a":1,"b":2}
print(pd.Series(a,index=["1","2"]))
print(pd.Series(a,index=["a","b"]))
```

<h3> finding and removing outliers </h3>

```python
def find_outliers(ds,col):
    quart1 = ds[col].quantile(0.25)
    quart2 = ds[col].quantile(0.75)
    IQR = quart2 - quart1
    low_val = quart1 - 1.5*IQR
    high_val = quart2 - 1.5*IQR
    ds = ds.loc[(ds[col] < low_val) | (ds[col] > high_val)]
    return ds
find_outliers(datasets,'acceleration')
```

```python
def remove_outliers(ds,col):
    quart1 = ds[col].quantile(0.25)
    quart2 = ds[col].quantile(0.75)
    IQR = quart2 - quart1
    low_val = quart1 - 1.5*IQR
    high_val = quart2 - 1.5*IQR
    ds = ds.loc[(ds[col] >= low_val) | (ds[col] <= high_val)]
    return ds
remove_outliers(datasets,'acceleration')
```
